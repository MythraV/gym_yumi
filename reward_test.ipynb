```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
distance = np.arange(0, 10, .001)

tested_reward = lambda d: 1 - 2*d
report_reward = lambda d: -2 * abs(.01 - 2 * d)
def modified_report_reward(d):
    if d <= .01:
        return 10
    else: 
        return report_reward(d)


plt.plot(distance, tested_reward(distance), label = "tested_reward")
plt.plot(distance, report_reward(distance), label = "report_rewrad")
plt.plot(distance, np.vectorize(modified_report_reward)(distance), label = "modified_report_rewrad")
plt.legend()
```
